```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
pwd=os.getcwd()
filepath=os.path.join(pwd,'Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv')
data_import=pd.read_csv(filepath)
data=data_import.copy()
#data.columns
```

```python
filtered_data = data[(data['TopicDesc'] == 'Cigarette Use (Adults)') & (data['MeasureDesc'] == 'Current Smoking')]
```

```python
grouped_data = filtered_data.groupby(['YEAR','LocationAbbr','LocationDesc','Gender', 'Race', 'Age','Education'])['Data_Value'].mean().reset_index()
```

```python
grouped_data[(grouped_data['Race']=='All Races')&((grouped_data['Age']=='All Ages'))].reset_index(drop=True)
```

```python
grouped_data.to_csv("Behavioural_risk_cigarette_use.csv",index=False)
```
