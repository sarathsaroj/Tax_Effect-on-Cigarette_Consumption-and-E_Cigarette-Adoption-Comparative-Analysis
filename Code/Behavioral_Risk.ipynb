```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
pwd=os.getcwd()
filepath=os.path.join(pwd,'Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv')
data_import=pd.read_csv(filepath)
data=data_import.copy()
#data.columns
```

```python
filtered_data = data[(data['TopicDesc'] == 'Cigarette Use (Adults)') & (data['MeasureDesc'] == 'Current Smoking')]
grouped_data_2011_2019 = filtered_data.groupby(['YEAR','LocationAbbr','LocationDesc','Gender', 'Race', 'Age','Education','GeoLocation'])['Data_Value'].mean().reset_index()
#grouped_data_2011_2019
```

```python
grouped_data_2011_2019.to_csv("Cigarette_Use_2011-2019.csv",index=False)
```

```python
Old_data=pd.read_csv('BRFSS_Prevalence_and_Trends_Data__Tobacco_Use_-_Four_Level_Smoking_Data_for_1995-2010.csv')
Old_data['Data_Value']=100-Old_data['Never smoked']
Old_data.rename(columns={"Year":'YEAR','State':'LocationDesc'},inplace=True)
Old_data[['Gender','Race','Age','Education']]=['Overall','All Races','All Ages','All Grades']
#Old_data
```

```python
New_data_overall=grouped_data_2011_2019[(grouped_data_2011_2019['Race']=='All Races')&(grouped_data_2011_2019['Age']=='All Ages')&(grouped_data_2011_2019['Education']=='All Grades')].reset_index(drop=True)
data_1995_2011=pd.concat([New_data_overall,Old_data], axis=0)
```

```python
data_1995_2011=pd.concat([New_data_overall,Old_data], axis=0)
```

```python
Loc_code=data_1995_2011.groupby(['LocationAbbr','LocationDesc'])['GeoLocation'].first().reset_index()
location_mapping=Loc_code.set_index('LocationDesc').to_dict()
```

```python
data_1995_2011['LocationAbbr'] = data_1995_2011['LocationAbbr'].fillna(data_1995_2011['LocationDesc'].map(location_mapping['LocationAbbr']))
data_1995_2011['GeoLocation']=data_1995_2011['GeoLocation'].fillna(data_1995_2011['LocationDesc'].map(location_mapping['GeoLocation']))
data_1995_2011.drop(columns=['Smoke everyday','Smoke some days','Former smoker','Never smoked','Counties','States','Location 1'])
```

```python
data_1995_2011.to_csv('Overall_Limited_data_1995-2019.csv')
```
